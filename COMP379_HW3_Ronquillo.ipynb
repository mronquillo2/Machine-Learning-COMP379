```python
titanic_knn = pd.DataFrame(titanic_scl, columns = ['Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived'])
titanic_knn
```

```python
x_train_knn, x_test_knn, y_train_knn, y_test_knn = train_test_split(x,y, test_size = .25, random_state=789)
x_train_knn
x_test_knn
y_train_knn
y_test_knn
```

```python
#knn
# 1) given a new point
# 2) get distances of all vectors from point
# 3) put all distances in a list
# 4f find the n nunmber of vectors closest to point from list
# 5) assign classification based on majority of classes


x_train_knn = np.array(x_train_knn)
x_train_knn2 = np.array(x_train_knn)
x_test_knn = np.array(x_test_knn)
y_train_knn = np.array(y_train_knn)
y_test_knn = np.array(y_test_knn)

```

```python
#k = 1
pred_class = []
for newpoint in x_test_knn:
  distances = []
  for i in x_train_knn:
    euclad =  np.sqrt(((newpoint[0]-i[0])**2) + 
                      ((newpoint[1]-i[1])**2) + 
                      ((newpoint[2]-i[2])**2) + 
                      ((newpoint[3]-i[3])**2) + 
                      ((newpoint[4]-i[4])**2) + 
                      ((newpoint[5]-i[5])**2))
    distances.append(euclad)
  mindist = min(distances)
  index = distances.index(mindist)
  #print(index)
  #print(x_train_knn[index])
  #print(y_train_knn[index])
  pred_class.append(y_train_knn[index])
    #find min distance in list
    #find index of row for that distance
    #use index to find class of that value in y
    #classify that point as the class of value from y



print(pred_class)
print(y_test_knn)

```

```python
print('Misclassified samples: %d' % (y_test_knn != pred_class).sum())
print('Accuracy = ' + str(float((y_test_knn == pred_class).sum()) / len(y_test_knn)*100)+'% with k = 1')
```

```python
k3_train_knn, k3_test_knn = train_test_split(titanic_scl, test_size = .25, random_state=789)
k3_train_knn
k3_test_knn
```

```python
k3_train_knn = np.array(k3_train_knn)
k3_train_knn2 = np.array(k3_train_knn)
k3_test_knn = np.array(k3_test_knn)

#print(k3_train_knn2)
#print(k3_train_knn2[:,6])
```

```python
k = 3
kpred_class = []
#iter = 0
for newpoint in k3_test_knn:
  #print("Iteration#" + str(iter))
  #iter += 1
  distances = []
  for i in k3_train_knn2:
    euclad =  np.sqrt(((newpoint[0]-i[0])**2) + 
                      ((newpoint[1]-i[1])**2) + 
                      ((newpoint[2]-i[2])**2) + 
                      ((newpoint[3]-i[3])**2) + 
                      ((newpoint[4]-i[4])**2) + 
                      ((newpoint[5]-i[5])**2))
    distances.append(euclad)
  dist = np.array(distances)
  arr_with_dist = np.insert(k3_train_knn2,7,dist,axis=1)
  sort_arr = arr_with_dist[arr_with_dist[:, 7].argsort()]
  #print(sort_arr)
  for classif in range(k):
    #print(sort_arr[:,6][classif])
    classif_lst = []
    classif_lst.append(sort_arr[:,6][classif])
    #print(classif_lst)
  majority = max(set(classif_lst), key = classif_lst.count)
  #print(majority)
  prediction = majority
  kpred_class.append(prediction)
  
  
  #print(arr_with_dist)
  #print(x_with_dist)
  #mindist = min(distances)
  #index = distances.index(mindist)
  #print(index)
  #print(x_train_knn[index])
  #print(y_train_knn[index])
  #pred_class.append(y_train_knn[index])
    #find min distance in list
    #find index of row for that distance
    #use index to find class of that value in y
    #classify that point as the class of value from y

#print(kpred_class)
#print(k3_test_knn[:,6])

print('Misclassified samples: %d' % (k3_test_knn[:, 6] != kpred_class).sum())
print('Accuracy = ' + str(float(((k3_test_knn[:, 6] == kpred_class).sum()) / len(k3_test_knn[:, 6])*100)) + '% with k = ' + str(k))
```
