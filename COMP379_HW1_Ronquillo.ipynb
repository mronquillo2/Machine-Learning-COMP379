```python
#import necessary libraries and packages
import pandas as pd
import numpy as np
```

```python
###for google colab only 
#read csv from file path after uploading to session storage
train = pd.read_csv("/content/train.csv", sep = ',')
train
```

```python
#exclude unnecessary variables: name, ticket, embarked, and cabin
  #keep survival column for reference, use pclass, sex, age, sibsp, parch, and fare
train2 = train[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
train2
```

```python
#Sex are strings, convert to binary for better data analysis
#set Female = 1, Male = 0
gender = pd.get_dummies(train2['Sex'])
#Gender
train2 = pd.concat((train2, gender), axis=1)
train2 = train2.drop(["Sex"], axis=1)
train2 = train2.drop(["male"], axis=1)
train3 = train2.rename(columns={"female": "Sex"})
train3
```

```python
#first general analysis of training set as a whole
  # the mean surival rate is .3838
    # more people died than survived in the dataset
  # mean Pclass is 2.308 with the 50th and 75th %ile being 3 and 25th %ile being 2
    # more people in 2nd and 3rd class than 1st class
  # mean age is 30 years old, with 20 year olds in 25th percentile and 38 year olds in 75th percentile
  # sibling/spouse had a mean of .5, max = 8
    # most people either had 1 sibling/spouse or none, at most someone had 8 sib/sp
  # mean parch is low, not very significant but still could be a deciding factor
  # mean fare was $32 with a max fare of $512, price range is pretty significant/wide
  # mean sex was .35
    # more males than females in dataset
train3.describe()
```

```python
#analyze statistics for those who survived in the data set
train_alive = train3[train3["Survived"] == 1]
train_alive.describe()
```

```python
#if most females survived, what factors did the males have to survive?
train_alive_male= train_alive[train_alive["Sex"] == 0]
train_alive_male.describe()
```

```python
#analyze statistics for those who did not survive in the data set
train_dead = train3[train3["Survived"] == 0]
train_dead.describe()
```

```python
#if most males, what factors did the females that died have?
train_dead_fem = train_dead[train_dead["Sex"] == 1]
train_dead_fem.describe()
```

```python
#Main takeaways
  # mean sex for survived was .68, mean sex for not survived was .14
    # most male did not survive, females whose pclass was 2 or 3 especially did not survive
  # mean pclass for survived was 1.95 25th %ile = 1 50th% %ile = 2
    # mean pclass not survive was 2.53, 25th %ile = 2 50th %ile = 3
      # those in 3rd class probably did not survive, 1st class likely survived
        # 2nd class depends on fare higher price = better placement in class
  # Fare for survived was a lot higher than did not survive
    # mean survived fare = $48, mean non survived fare = $22
    # both the 50th %ile for suvived and 75th %ile for non survived was $26
      # use this overlap for condition to determine survived or not
    # infer that the more you payed the higher class/ better placement
  
  # since the statistics for age were roughly the same between those who survived and those who did not, 
    # placing a condition on age is not very helpful
    # this also applies to Sibsp and Parch
      # both had no real difference in means except 75th %ile for parch for those who survived had at most 1 

# conditions
  # if person is in 2nd or 3rd class AND fare was less than $26, probably dead
    # most males died, and females mostly died if they were in 3rd class (based on mean pclass)
  # if person is in 1st or 2nd class AND fare was more than $26, alive
    # pclass and fare big deciding factor, the higher class and more expensive fare probably meant better placement in boat
  # since more of the survivors were female and previous condition already filters out dead ones, the rest are probably alive
```

```python
#initialize empty set to put all predictions (yhat) in
#create a loop that iterates through every row
  #if values in each column meet specific conditions:
    #assign a 1 for survived and 0 for non-survived
      #else, pass through other conditions and assign 1 or 0 
    #append the 1 or 0 to a list called yhat
      #yhat will contain all the predictions of survival based on conditions, will use to compare to actual survived column


yhat_train = []
for i in train3.index:
  if train3["Pclass"][i] >= 2 and train3["Fare"][i] < 26:
    srv = 0
    yhat_train.append(srv)
  elif train3["Sex"][i] == 0 and train3["Pclass"][i] <= 2 and train3["Fare"][i] > 26: #and train3["SibSp"][i] == 0:
    srv = 1
    yhat_train.append(srv)
  else:
    srv = 1
    yhat_train.append(srv)
#print(yhat)
```

```python
#combine "survived" column from training data with yhat predicted values from algorithm with conditions to compare
actual_train = np.array(train3['Survived'])
compare_train = pd.DataFrame({"yhat": yhat_train, 'actual': actual_train})
compare_train
```

```python
#compare amount of yhat values of alive to amount of actual values of alive
yhat_alive =  compare_train[compare_train['yhat'] == 1]
print("Predicted alive in training set: " + str(len(yhat_alive)))
actual_alive =  compare_train[compare_train['actual'] == 1]
print("Actual alive in training set: " + str(len(actual_alive)))
```

```python
#compare amount of yhat values of dead to amount of actual values of dead
yhat_dead =  compare_train[compare_train['yhat'] == 0]
print("Predicted alive in training set: " + str(len(yhat_dead)))
actual_dead =  compare_train[compare_train['actual'] == 0]
print("Actual alive in training set: " + str(len(actual_dead)))
```

```python
#read in test data csv using same method as training
test = pd.read_csv('/content/test.csv', sep = ',')
gender = pd.read_csv('/content/gender_submission.csv', sep = ',')
```

```python
#test data does not have "survived" column, concatinate gender_submissions data with test on "PassengerID" 
test2 = pd.merge(gender, test, how = 'inner', on = 'PassengerId')
#create dataframe of test data using only "pclass", "sex", and "fare"
test3 = test2[['Survived', 'Pclass', 'Sex', 'Fare']]
test3
```

```python
#convert from male/female to 0/1
gendertst = pd.get_dummies(test3['Sex'])
test3 = pd.concat((test3, gendertst), axis=1)
test3 = test3.drop(["Sex"], axis=1)
test3 = test3.drop(["male"], axis=1)
test3 = test3.rename(columns={"female": "Sex"})
test3
```

```python
#apply same algorithm from training to testing 
yhattst = []
for i in test3.index:
  if test3["Pclass"][i] >= 2 and test3["Fare"][i] < 26:
    srvtst = 0
    yhattst.append(srvtst)
  elif test3["Sex"][i] == 0 and test3["Pclass"][i] <= 2 and test3["Fare"][i] >26: #and test["SibSp"][i] == 0::
    srvtst = 1
    yhattst.append(srvtst)
  else:
    srvtst = 1
    yhattst.append(srvtst)
print(yhattst)
```

```python
#combine "survived" column from training data with yhat predicted values from algorithm with conditions to compare for test
actualtst = np.array(test3['Survived'])
comparetst = pd.DataFrame({"yhattst": yhattst, 'actualtst': actualtst})
comparetst
```

```python
#compare amount of yhat values of alive to amount of actual values of alive for test
yhattst_alive = comparetst[comparetst['yhattst'] == 1]
actualtst_alive =  comparetst[comparetst['actualtst'] == 1]
print("Yhat prediction of survivors: " + str(len(yhattst_alive)))
print("Actual prediction of survivors: " + str(len(actualtst_alive)))
```

```python
#compare amount of yhat values of dead to amount of actual values of dead for test
yhattst_dead = comparetst[comparetst['yhattst'] == 0]
actualtst_dead =  comparetst[comparetst['actualtst'] == 0]
print("Yhat prediction of non-survivors: " + str(len(yhattst_dead)))
print("Actual prediction of non-survivors: " + str(len(actualtst_dead)))
```
